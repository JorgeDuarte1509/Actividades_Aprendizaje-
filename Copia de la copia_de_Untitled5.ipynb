```python
import pandas as pd
import numpy as np
import seaborn as sns
Payments = 'https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv'
```

```python
Payments_df = pd.read_csv(Payments)
df = Payments_df.copy()
```

```python
df.dropna(inplace=True) # Eliminamos los renglones que no tnegan dato en alguna columna
ndf = df[df['X3']>0] # Asigna al nuevo dataframe 2 valores arriba de 0 de la columna "X3" (Nos elimina valores con 0)
ndf = ndf[ndf['X3']<=4] # Asigna al nuevo dataframe 2 valores iguales o menores a 4 de la columna "X3" (Nos elimina valores mayores a 4)
ndf = ndf[ndf['X4']>0] # Asigna al nuevo dataframe 2 valores arriba de 0 de la columna "X4" (Nos elimina valores con 0)
ndf # Enseña el nuevo dataframe 2 con los datos limpios
```

```python
ndf2 = ndf[['X12','X13','X14','X15','X16','X17']]
ndf2.corr()
```

```python
Varience = np.cov([ndf2.X12, ndf2.X13, ndf2.X14, ndf2.X15, ndf2.X16, ndf2.X17])
Varience
```

```python
ndf2.var().sum()
```

```python
t= ndf2.var().sum()
varX12 = Varience[0][0]
varX13 = Varience[1][1]
varX14 = Varience[2][2]
varX15 = Varience[3][3]
varX16 = Varience[4][4]
varX17 = Varience[5][5]
```

```python
print('Total Varience \t\t\t:', t.round(2) )
print('Varience prop. of X12 \t:', ((varX12/t)*100).round(2),'%' )
print('Varience prop. of X13 \t:', ((varX13/t)*100).round(2),'%' )
print('Varience prop. of X14 \t:', ((varX14/t)*100).round(2),'%' )
print('Varience prop. of X15 \t:', ((varX15/t)*100).round(2),'%' )
print('Varience prop. of X16 \t:', ((varX16/t)*100).round(2),'%' )
print('Varience prop. of X17 \t:', ((varX17/t)*100).round(2),'%' )
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pcs = PCA(n_components=6)

pcs.fit(ndf2[['X12','X13','X14','X15','X16','X17']])

pcsSummary = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                          'Proportion of variance': pcs.explained_variance_ratio_,
                          'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)#column 
                          }
                          )
pcsSummary
```

```python
pcsSummary = pcsSummary.transpose()
pcsSummary.round(2)
```

```python
pcsSummary.columns = ['PC1', 'PC2','PC3', 'PC4','PC5', 'PC6']
pcsSummary = pcsSummary.round(4)
pcsSummary
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
PC_components = np.arange(pcs.n_components_) + 1
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
